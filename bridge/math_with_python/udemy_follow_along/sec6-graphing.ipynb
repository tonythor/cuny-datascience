```python
# import required packages at the top of the script!
import sympy as sym
import numpy as np
from IPython.display import display, Math
import matplotlib.pyplot as plt
```

```python
x=3
y=5
plt.plot(x,y,'r+', label='MyChart')
plt.axis('square')
plt.axis([-6,+6,-6,+6])
plt.grid()
plt.legend()
## adjusting the axis
axis = plt.gca()
ylim = axis.get_ylim()
axis.set_ylim(ylim[0],10)

plt.show()

```

```python
x = sym.symbols('x')
formula=x**2-3*x
display(Math(sym.latex(formula)))

for r in range(-10,12):
    l,r = (r, formula.subs(x,r))
    plt.plot(l,r, "r+")
plt.xlabel('X')
plt.ylabel(f'f(x) {formula}')
plt.grid()
plt.show()



         

```
